```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-poster')
%matplotlib inline
```

```python
# sampling rate
sr = 2000 # This means we are collecting 2000 samples every second.

# sampling interval
ts = 1.0/sr # delta time between two sample.
t = np.arange(0,1,ts)

freq = 1.
x = 3*np.sin(2*np.pi*freq*t)

freq = 4
x += np.sin(2*np.pi*freq*t)

freq = 7   
x += 0.5* np.sin(2*np.pi*freq*t)

freq = 8   
x += 0.5* np.sin(2*np.pi*freq*t)

plt.figure(figsize = (8, 6))
plt.plot(t, x, 'r')
plt.ylabel('Amplitude')
```

```python
from numpy.fft import fft, ifft

X = fft(x)
```

```python
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T 
```

```python
plt.stem(freq, np.abs(X), 'b', markerfmt=" ", basefmt="-b")
plt.xlim(0, 10)
```

<font color='green'>In above figure, we see frequency components of 1 Hz, 4 Hz, 7 Hz and 8 Hz. This is because the input signal that we provided had these frequencies in them.</font>
